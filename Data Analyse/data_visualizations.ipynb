```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
```

```python
import pandas as pd

# Load the prepared data from the CSV file
df = pd.read_csv('prepared_data.csv')

# Display the first few rows to verify
df.head()
```

# Word Cloud for Titles and Content

```python
# Combine all titles into a single string
all_titles = ' '.join(df['title'].dropna())

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_titles)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Titles')
plt.show()
```

```python
# Combine all titles into a single string
all_contents = ' '.join(df['content'].dropna())

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_contents)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Titles')
plt.show()
```

# Bar Chart of Most Frequent Words

```python
# Split and count words in titles
all_words = ' '.join(df['title'].dropna()).split()
word_counts = Counter(all_words)

# Get the top 10 most common words
top_words = word_counts.most_common(10)

# Plot the bar chart
words, counts = zip(*top_words)
plt.figure(figsize=(10, 5))
plt.bar(words, counts, color='skyblue')
plt.title('Top 10 Most Frequent Words in Titles')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Split and count words in titles
all_words = ' '.join(df['content'].dropna()).split()
word_counts = Counter(all_words)

# Get the top 10 most common words
top_words = word_counts.most_common(10)

# Plot the bar chart
words, counts = zip(*top_words)
plt.figure(figsize=(10, 5))
plt.bar(words, counts, color='skyblue')
plt.title('Top 10 Most Frequent Words in Titles')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

# Length Distribution of Titles and Content

```python
# Calculate the length of each title and content
df['title_length'] = df['title'].apply(lambda x: len(x.split()))
df['content_length'] = df['content'].apply(lambda x: len(x.split()))

# Plot the distribution of title lengths
plt.figure(figsize=(10, 5))
plt.hist(df['title_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Title Lengths (Word Count)')
plt.xlabel('Title Length (Words)')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of content lengths
plt.figure(figsize=(10, 5))
plt.hist(df['content_length'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Content Lengths (Word Count)')
plt.xlabel('Content Length (Words)')
plt.ylabel('Frequency')
plt.show()
```
