```python
import pandas as pd
import numpy as np
import os
```

# data preparing

```python
# Define the directory where your datasets are stored
data_dir = '/home/am44/A/ENSAK/S2/Machine Learning/project/dataset/all_data_sets'

# List all files in the directory
all_files = os.listdir(data_dir)

# Filter out non-CSV files if necessary
csv_files = [f for f in all_files if f.endswith('.csv')]

# Load all CSV files into a list of DataFrames
df_list = [pd.read_csv(os.path.join(data_dir, f)) for f in csv_files]
```

```python
# Remove DataFrames with index 0 and 1
df_list = df_list[2:]  # Keep only DataFrames starting from index 2
```

```python
# Drop columns from the DataFrame at index 2 (originally index 4)
df_list[2] = df_list[2].drop(columns=['url', 'authors', 'timestamp', 'tags'])
```

```python
# Drop columns from the DataFrame at index 1 (originally index 3)
df_list[1] = df_list[1].drop(columns=['author', 'claps', 'reading_time', 'link'])
```

```python
# Step 1: Extract and rename columns from each DataFrame
# DataFrame at index 0
df0 = df_list[0][['Title', 'Text']].rename(columns={'Title': 'title', 'Text': 'content'})

# DataFrame at index 1
df1 = df_list[1][['title', 'text']].rename(columns={'text': 'content'})

# DataFrame at index 2
df2 = df_list[2][['title', 'text']].rename(columns={'text': 'content'})

# Step 2: Combine the DataFrames into one
combined_df = pd.concat([df0, df1, df2], ignore_index=True)

# Step 3: Verify the new DataFrame
combined_df.head()
```

```python
combined_df.info()
```

```python
combined_df.isnull().sum()
```

```python
# Fill empty cells in the 'title' column with the word "Title"
combined_df['title'] = combined_df['title'].fillna('Title')

# Verify the changes
combined_df['title'].isnull().sum()
```

```python
combined_df.describe()
```

```python
# Drop duplicated rows
combined_df = combined_df.drop_duplicates()
```
